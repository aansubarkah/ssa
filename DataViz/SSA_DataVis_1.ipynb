# Membuat Line Chart

## Contoh Nilai Penjualan Harian

```python
import matplotlib.dates as mdates
```

```python
df['ModifiedDate'] = pd.to_datetime(df['ModifiedDate'])

daily_sales = df.groupby('ModifiedDate')['LineTotal'].sum().reset_index()
daily_sales = daily_sales.sort_values('ModifiedDate')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(daily_sales['ModifiedDate'], daily_sales['LineTotal'], linewidth=2, color='blue')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(daily_sales['ModifiedDate'], daily_sales['LineTotal'], linewidth=2, color='blue')

# Tambah Judul dan Label
ax.set_title('Daily Total Order Quantity Over Time', fontsize=16)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Total Order Quantity', fontsize=12)

# Format x-axis agar tanggal terlihat rapi
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45, ha='right')

# Tambah grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Tandai titik tertinggi dan terendah
max_point = daily_sales.loc[daily_sales['LineTotal'].idxmax()]
min_point = daily_sales.loc[daily_sales['LineTotal'].idxmin()]

ax.annotate(f'Max: {max_point["LineTotal"]}', 
          xy=(max_point['ModifiedDate'], max_point['LineTotal']),
          xytext=(10, 10), textcoords='offset points',
          arrowprops=dict(arrowstyle='->'))

ax.annotate(f'Min: {min_point["LineTotal"]}', 
          xy=(min_point['ModifiedDate'], min_point['LineTotal']),
          xytext=(10, -10), textcoords='offset points',
          arrowprops=dict(arrowstyle='->'))

# Adjust layout
plt.tight_layout()

plt.show()
```

## Contoh Jumlah Penjualan Harian untuk tahun 2012

```python
daily_sales_count = df[df['ModifiedDate'].dt.year == 2012].groupby('ModifiedDate')['SalesOrderID'].nunique().reset_index()
daily_sales_count = daily_sales_count.sort_values('ModifiedDate')
daily_sales_count.columns = ['ModifiedDate', 'OrderCount']
```

```python
daily_sales_count
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(daily_sales_count['ModifiedDate'], daily_sales_count['OrderCount'], linewidth=2, color='blue')

# Tambah Judul dan Label
ax.set_title('Jumlah Penjualan Harian pada 2012', fontsize=16)
ax.set_xlabel('Tanggal', fontsize=12)
ax.set_ylabel('Jumlah Penjualan', fontsize=12)

# Format x-axis agar tanggal terlihat rapi
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45, ha='right')

# Tambah grid lines
ax.grid(True, linestyle='--', alpha=0.7)

# Tandai titik tertinggi dan terendah
max_point = daily_sales_count.loc[daily_sales_count['OrderCount'].idxmax()]
min_point = daily_sales_count.loc[daily_sales_count['OrderCount'].idxmin()]

ax.annotate(f'Tertinggi: {max_point["OrderCount"]}', 
          xy=(max_point['ModifiedDate'], max_point['OrderCount']),
          xytext=(10, 10), textcoords='offset points',
          arrowprops=dict(arrowstyle='->'))

ax.annotate(f'Terendah: {min_point["OrderCount"]}', 
          xy=(min_point['ModifiedDate'], min_point['OrderCount']),
          xytext=(10, -10), textcoords='offset points',
          arrowprops=dict(arrowstyle='->'))

# Adjust layout
plt.tight_layout()

plt.show()
```
