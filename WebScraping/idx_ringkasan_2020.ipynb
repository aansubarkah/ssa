# Buat Tanggal

```python
start_date = datetime(2020, 1, 1)
end_date = datetime(2021, 1, 1)
#end_date = datetime.today()
```

```python
url_start = 'https://www.idx.co.id/primary/TradingSummary/GetIndexSummary?date='
url_end = '&length=9999&start=0'

current_date = start_date
while current_date <= end_date:
    try:
        cur_date = current_date.strftime('%Y%m%d')
        print(cur_date)

        url_to = url_start + cur_date + url_end
        driver.get(url_to)

        teks = driver.find_element(By.TAG_NAME, 'pre')
        data_in_json = json.loads(teks.text)
        with open(f'{new_dir}/{cur_date}.json', 'w') as f:
            json.dump(data_in_json['data'], f, indent=4)

        #r = S.get(url_to, headers=headers)
        #if r.status_code == 200:
            #with open(f'{new_dir}/{cur_date}.json', 'w') as f:
                #json.dump(r.json(), f)

        #else:
            #print('Error, status code', cur_date)
        time.sleep(1)
    except:
        print('Error', cur_date)

    current_date += timedelta(days=1)
```

# Audit JSON

## Apa ada Tanggal yang belum didownload

```python
new_dir = 'idx_data_2020'
```

```python
start_date = datetime(2020, 1, 1)
#end_date = datetime(2021, 1, 1)
end_date = datetime.today()
```

```python
json_files = [f.replace('.json', '') for f in os.listdir(new_dir) if f.endswith('.json')]
```

```python
#json_files
```

```python
url_start = 'https://www.idx.co.id/primary/TradingSummary/GetIndexSummary?date='
url_end = '&length=9999&start=0'

current_date = start_date
while current_date <= end_date:
    cur_date = current_date.strftime('%Y%m%d')
    if cur_date not in json_files:
        print(cur_date)
        url_to = url_start + cur_date + url_end
        driver.get(url_to)

        teks = driver.find_element(By.TAG_NAME, 'pre')
        data_in_json = json.loads(teks.text)
        with open(f'{new_dir}/{cur_date}.json', 'w') as f:
            json.dump(data_in_json['data'], f, indent=4)

        time.sleep(1)
    current_date += timedelta(days=1)
```

# Concat the data

```python
data_ls = []

for file in os.listdir(new_dir):
    if file.endswith('json'):
        print(file)

        with open(os.path.join(new_dir, file), 'r') as f:
            data = json.load(f)
            df_temp = pd.DataFrame(data)
            data_ls.append(df_temp)
            
            #data_ls.append(pd.DataFrame(data))
            
            
        #print(data)
df = pd.concat(data_ls)
```
