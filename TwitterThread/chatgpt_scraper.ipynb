```python
import requests
from bs4 import BeautifulSoup

# URL halaman Wikipedia yang ingin di-scrape
url = 'https://id.wikipedia.org/wiki/Python_(bahasa_pemrograman)'

# Mengirim permintaan HTTP ke URL
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    # Parsing konten halaman dengan BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Mengambil judul artikel
    title = soup.find('h1', id='firstHeading').text
    print(f"Judul Artikel: {title}")
    
    # Mengambil semua paragraf dalam artikel
    paragraphs = soup.find_all('p')
    for i, para in enumerate(paragraphs):
        print(f"Paragraf {i+1}: {para.text}\n")
else:
    print("Gagal mengakses halaman Wikipedia")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL halaman Wikipedia yang ingin di-scrape
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

# Mengirim permintaan HTTP ke URL
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    # Parsing konten halaman dengan BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Menemukan tabel yang berisi daftar perusahaan S&P 500
    table = soup.find('table', {'id': 'constituents'})
    
    # Mengambil header kolom
    headers = []
    for th in table.find_all('th'):
        headers.append(th.text.strip())
    
    # Mengambil data dari setiap baris di tabel
    rows = []
    for tr in table.find_all('tr')[1:]:
        cells = tr.find_all('td')
        row = [cell.text.strip() for cell in cells]
        rows.append(row)
    
    # Membuat DataFrame dari data yang diambil
    df = pd.DataFrame(rows, columns=headers)
    
    # Menampilkan DataFrame
    print(df.head())
    
    # Menyimpan data ke file CSV
    df.to_csv('sp500_companies.csv', index=False)
    print("Data telah disimpan ke 'sp500_companies.csv'")
    
else:
    print("Gagal mengakses halaman Wikipedia")

```

```python
import requests
from bs4 import BeautifulSoup

# URL halaman utama Wikipedia Indonesia
url = 'https://id.wikipedia.org/wiki/Halaman_Utama'

# Mengirim permintaan HTTP ke URL
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    # Parsing konten halaman dengan BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Mencari bagian "Artikel pilihan" di halaman
    artikel_pilihan = soup.find('div', {'id': 'mf-artikelpilihan'})
    
    # Mengambil judul artikel
    judul = artikel_pilihan.find('b').text
    
    # Mengambil deskripsi singkat artikel
    deskripsi = artikel_pilihan.find('p').text
    
    # Mengambil URL artikel
    artikel_url = artikel_pilihan.find('b').find('a')['href']
    artikel_url_full = f"https://id.wikipedia.org{artikel_url}"
    
    # Menampilkan hasil scraping
    print(f"Judul Artikel Pilihan: {judul}")
    print(f"Deskripsi: {deskripsi}")
    print(f"URL: {artikel_url_full}")
    
else:
    print("Gagal mengakses halaman Wikipedia")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL halaman Ringkasan Indeks IDX
url = 'https://www.idx.co.id/id/data-pasar/ringkasan-perdagangan/ringkasan-indeks'

# Mengirim permintaan HTTP ke URL
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    # Parsing konten halaman dengan BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Mencari tabel yang berisi data historikal
    table = soup.find('table', {'class': 'table table-striped table-hover table-bordered table-condensed'})
    
    # Mengambil header kolom
    headers = []
    for th in table.find_all('th'):
        headers.append(th.text.strip())
    
    # Mengambil data dari setiap baris di tabel
    rows = []
    for tr in table.find_all('tr')[1:]:
        cells = tr.find_all('td')
        row = [cell.text.strip() for cell in cells]
        rows.append(row)
    
    # Membuat DataFrame dari data yang diambil
    df = pd.DataFrame(rows, columns=headers)
    
    # Menampilkan DataFrame
    print(df.head())
    
    # Menyimpan data ke file CSV
    df.to_csv('ringkasan_indeks.csv', index=False)
    print("Data telah disimpan ke 'ringkasan_indeks.csv'")
    
else:
    print("Gagal mengakses halaman IDX")

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# Path ke WebDriver, misalnya untuk Chrome
driver_path = '/path/to/chromedriver'  # Ganti dengan path ke WebDriver Anda
url = 'https://www.idx.co.id/id/data-pasar/ringkasan-perdagangan/ringkasan-indeks/'

# Membuka browser dengan Selenium
#driver = webdriver.Chrome(executable_path=driver_path)
driver.get(url)

# Tunggu beberapa saat hingga halaman dimuat
time.sleep(5)

# Menemukan tabel di halaman
table = driver.find_element(By.CLASS_NAME, 'table-condensed')

# Mengambil header kolom dari tabel
headers = []
for header in table.find_elements(By.TAG_NAME, 'th'):
    headers.append(header.text.strip())

# Mengambil data dari setiap baris di tabel
rows = []
for row in table.find_elements(By.TAG_NAME, 'tr')[1:]:
    cells = row.find_elements(By.TAG_NAME, 'td')
    row_data = [cell.text.strip() for cell in cells]
    rows.append(row_data)

# Membuat DataFrame dari data yang diambil
df = pd.DataFrame(rows, columns=headers)

# Menampilkan DataFrame
print(df.head())

# Menyimpan data ke file CSV
df.to_csv('ringkasan_indeks_selenium.csv', index=False)
print("Data telah disimpan ke 'ringkasan_indeks_selenium.csv'")

# Menutup browser
driver.quit()
```
