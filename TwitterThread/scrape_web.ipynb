```python
options = Options()
#options.headless = False
options.add_argument('--headless=new')
#driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)
driver = webdriver.Chrome(options=options)
```

```python
driver.get('https://www.idx.co.id/primary/TradingSummary/GetIndexSummary?date=20240821&length=9999&start=0')
```

```python
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)

cur_date = start_date
```

```python
while cur_date <= end_date:
    print(cur_date.strftime('%Y%m%d'))
    cur_date = cur_date + timedelta(days=1)
```

```python
start_date = datetime(2020, 5, 11)
end_date = datetime(2024, 8, 22)

cur_date = start_date
```

```python
#pre_link = 'https://www.idx.co.id/primary/TradingSummary/GetIndexSummary?date=20240821&length=9999&start=0'
pre_link = 'https://www.idx.co.id/primary/TradingSummary/GetIndexSummary?date='
post_link = '&length=9999&start=0'
while cur_date <= end_date:
    print(cur_date.strftime('%Y%m%d'))
    cur_link = pre_link + cur_date.strftime('%Y%m%d') + post_link

    driver.get(cur_link)
    element = driver.find_element(By.TAG_NAME, 'pre')
    datum = driver.page_source
    print(json.loads(element.text)['recordsTotal'])

    with open(cur_date.strftime('%Y%m%d') + '.json', 'w', encoding='utf-8') as f:
        json.dump(json.loads(element.text), f, ensure_ascii=False, indent=4)

    cur_date = cur_date + timedelta(days=1)
    time.sleep(3)
```
